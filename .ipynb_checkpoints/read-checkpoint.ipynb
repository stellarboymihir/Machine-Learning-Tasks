```python
from PIL import Image
import os

folder_path = '/home/mihir/Documents/Ml Project/Database/Bagvan Mohzaid'
images = []

for filename in os.listdir(folder_path):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        filepath = os.path.join(folder_path, filename)
        image = Image.open(filepath)
        images.append(image)

```

```python
from PIL import Image
import os, os.path

imgs = []
path = '/home/mihir/Documents/Ml Project/Database/Bagvan Mohzaid'
valid_images = [".jpg",".gif",".png",".tga"]
for f in os.listdir(path):
    ext = os.path.splitext(f)[1]
    if ext.lower() not in valid_images:
        continue
    imgs.append(Image.open(os.path.join(path,f)))

print(imgs)
```

This code is for reading and saving in another folder...

```python
from PIL import Image
import os

folder_path = '/home/mihir/Documents/Ml Project/Database/Bagvan Zaid1/'
output_path = '/home/mihir/Documents/Ml Project/Enhancement/abc'

for filename in os.listdir(folder_path):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        filepath = os.path.join(folder_path, filename)
        image = Image.open(filepath)
        output_file = os.path.join(output_path, filename)
        image.save(output_file)
```
